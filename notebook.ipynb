# Don't Kill The patient 

### As a doctor, you need to prescribe the best medicine to the patient after determining the type of disease and measuring its severity.
### After the diagnosis, you will first look up the best drug that cures the sickness from a given dictionary.
### Then you have to calculate the word score of each medicine in the list and return the most optimal drug.
<br>

### PS : The word score is calculated by adding the letters that constitute the medication name as follows:
### a=1, b=2, c=3, d=4, â€¦, z=26.


# Example : 
patient : "headache", 68

Medicaments = {"headache" : ['Analgon','Dolipran'] , 'stomach ache' : ['Aztreonam'..]}

1st step :

    we will get all the medicines for the headache : ['Analgon','Dolipran']
    
2nd step :
    Calculate the word score for each item in the list 
    
    Analgon   64 : a+n+a+l+g+o+n
    
    Dolipran  89 : d+o+l+i+p+r+a+n
    
3rd step :

    return closest one :
    
    Analgan is closer to 68
    

```python
Medicines = {"headache" : ['Analgon','Dolipran','aspirin','naproxen'] , 'stomach ache' : ['Aztreonam','Famotidine','Loperamide','Dipep'],
              'throat ache' : ['Acetaminophen','Dyclonine','Chloraseptic', 'Clamoxyl']}
```

```python
disease = input("the disease: ")
severity = int(input("severity: "))
if disease in Medicines:
    medicament = Medicines.get(disease)
    medicament_array = dict.fromkeys(medicament, 0)
    print(medicament_array)
    for x in medicament_array:
        for i in x:
            medicament_array[x] += ord(i.upper()) - 64
    print(disease, ' : ', severity)
    print(medicament_array)

    min_diff = severity
    best_medicament = ""
    for y in medicament_array:
        diff = abs(medicament_array.get(y) - severity)
        if (min_diff > diff):
            min_diff = diff
            best_medicament = y

    print(best_medicament)

else:
    print("n'existe pas")
```
